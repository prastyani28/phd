```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(r'C:\Users\erinap\OneDrive - Universitetet i Oslo\Documents\Experiments\GFZ\27.10.2022\MTS\CM1_EP_CP10\CM1_EP_CP10_python.xlsx')
# print(df)
```

```python
df.plot(x='Time',y='Differential stress')
plt.title('Time vs Differential stress')
plt.xlabel('Time (s)')
plt.ylabel('Differential stress (MPa)')
```

```python
df.plot(x='Axial strain', y='Differential stress')
plt.title('Axial strain vs Differential stress')
plt.xlabel('Axial strain')
plt.ylabel('Differential stress (MPa)')
```

```python
df.plot(x='Time',y=['SGV3','SGV4','SGH3','SGF1'])
plt.title('Strain gauges')
plt.xlabel('Time (s)')
plt.ylabel('Strain gauges')
```

```python
df.plot(x='Volumetric strain SGV3-SGH3',y='Differential stress')
plt.title('Volumetric strain vs Differential stress')
plt.xlabel('Volumetric strain')
plt.ylabel('Differential stress (MPa)')
```

```python
ndf = df[(df['Time'] < 9300) & (df['Time'] > 625)]
ndf.plot(x='Time',y='Differential stress')
plt.title('Time vs Differential stress after removing some part of the unloading part')
plt.xlabel('Time (s)')
plt.ylabel('Differential stress (MPa)')
```

```python
ndf.plot(x='Axial strain',y='Differential stress')
plt.title('Axial strain vs Differential stress')
plt.xlabel('Axial strain')
plt.ylabel('Differential stress (MPa)')
```

```python
ndf = df[(df['Axial strain'] < 0.003) & (df['Axial strain'] > 0.0022)]
ndf.plot(x='Axial strain',y='Differential stress')
plt.title('Young modulus')
plt.xlabel('Axial strain')
plt.ylabel('Differential stress (MPa)')
```

```python
import numpy as np
fig, ax = plt.subplots()
ax.set_ylabel('Young Moduli [MPa]')
ax.set_xlabel('Axial Strain []')
# ax.plot(ndf_mts['Axial strain'],ndf_mts['Differential stress']/ndf_mts['Axial strain'],'bo')
N = 100
axsmoo = np.convolve(ndf['Axial strain'], np.ones(N)/N, mode='same')
stsmoo = np.convolve(ndf['Differential stress'], np.ones(N)/N, mode='same')
ax.plot(ndf['Axial strain'],stsmoo/axsmoo,'bo')
# ax.plot(ndf_mts['Axial strain'],ndf_mts['Differential stress'],'bo')
```
