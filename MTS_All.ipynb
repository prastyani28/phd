```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load dataframe

df_mts_10mpa = pd.read_excel(r'C:\Users\erinap\OneDrive - Universitetet i Oslo\Documents\Experiments\GFZ\27.10.2022\MTS\CM1_EP_CP10\MTS_10MPa_new.xlsx')
df_mts_40mpa = pd.read_excel(r'C:\Users\erinap\OneDrive - Universitetet i Oslo\Documents\Experiments\GFZ\11.11.2022\MTS\MTS_40MPa_new.xlsx')
df_mts_100mpa = pd.read_excel(r'C:\Users\erinap\OneDrive - Universitetet i Oslo\Documents\Experiments\GFZ\08.11.2022\MTS\MTS_100MPa_new.xlsx')
```
